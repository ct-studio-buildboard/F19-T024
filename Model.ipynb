```python
import gensim
# Load Google's pre-trained Word2Vec model.
model = gensim.models.KeyedVectors.load_word2vec_format('GoogleNews-vectors-negative300.bin', binary=True)  
```

```python
word_vectors = model.wv
refined = data[data['query'].isin(word_vectors.vocab)]
```

# The following function runs the model - requires model training

```python

def list_ideas (word):
    sim_values = []
    for item in medioids_clust:
        sim_values.append(model.similarity(item, word))
    index = sim_values.index(max(sim_values))
    print(index)
    print(max(sim_values))
    return index

```
